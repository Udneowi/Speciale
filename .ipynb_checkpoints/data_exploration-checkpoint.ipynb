# Baselines

### Guessing on the most common

```python
data_hist_1 = []
for pep in good_peps_train:
    data = load_data([pep],1,"all_relabeled",stop_t=False)
    if len(data) == 0:
        continue
    data = data[0].t().reshape(-1).cpu().numpy()[1:]
    if len(data) <= 1000:
        continue
    ii_train = int(len(data)//(10/9))
    data_train = data[:ii_train]
    data_test = data[ii_train:]
    
    count = Counter(data_train)
    boo_1 = data_test == count.most_common(1)[0][0]
    accu = sum(boo_1)/len(data_test)
    data_hist_1.append(accu)
```

```python
bins = plt.hist(data_hist_1,bins = 50)
print(f"Mean accuracy when guessing on most common: {np.mean(data_hist_1)}")
```

### Most common but not self

```python
data_hist_2 = []
for pep in good_peps_train:
    data = load_data([pep],bs=1,dat_type="all_relabeled", stop_t=False)[0][1:]
    if len(data) == 0:
        continue
    data = data.t().reshape(-1).cpu().numpy()
    if len(data) <= 1000:
        continue
    ii_train = int(len(data)//(10/9))
    data_train = data[:ii_train]
    data_test = data[ii_train:]
    count = Counter(data_train,sorted =True)
    boo_1 = data_test == count.most_common(1)[0][0]
    boo_2 = (data_test[1:])[boo_1[:-1]] == count.most_common(2)[1][0]
    accu = (sum(boo_1)+sum(boo_2))/len(data_test)
    data_hist_2.append(accu)
```

```python
bins = plt.hist(data_hist_2,bins = 50)
print(f"Mean accuracy when guessing on most common: {np.mean(data_hist_2)}")
```

### Markov order 1

```python
data_hist_3 = []
temp = 0
n = 0 # threshold of amount of times a location must be visited before not guessing most common
for pep in good_peps_train:
    data = load_data([pep],bs=1,dat_type="all_relabeled")[0]
    if len(data) == 0:
        continue
    data = data.t().reshape(-1).cpu().numpy()[1:]
    if len(data) <= 1000:
        continue
    ii_train = int(len(data)//(10/9))
    data_train = data[:ii_train]
    data_test = data[ii_train:]
    count = Counter(data_train)
    
    count_dict = {}
    accu = 0 
    for i, key in enumerate(data_train[:-1]):
        if key in count_dict:
            if data_train[i+1] in count_dict[key]:
                count_dict[key][data_train[i+1]]+=1
            else:
                count_dict[key][data_train[i+1]]=1
        else:
            count_dict[key] = {data_train[i+1]:1}
    for key, value in count_dict.items():
        if count[key]<=n:
            count_dict[key] = count.most_common(1)[0][0]
        else:
            count_dict[key] = max(count_dict[key].items(), key=operator.itemgetter(1))[0]

    for i, key in enumerate(data_test[:-1]):
        if key in count_dict:
            if count_dict[key] == data_test[i+1]:
                accu+=1
        else:
            if count.most_common(1)[0][0]== data_test[i+1]:
                accu+=1
    data_hist_3.append(accu/len(data_test))
```

```python
bins = plt.hist(data_hist_3,bins = 50)
print(f"Mean accuracy when guessing on most common given previous location: {np.mean(data_hist_3)}")
```

### Guessing on most common given current location (Markov order 1), but guessing on most common if location not visited for more than N times.

```python
data_hist_5 = []
temp = 0
n = 2 # threshold of amount of times a location must be visited before not guessing most common
for pep in good_peps_train:
    data = load_data([pep],bs=1,dat_type="all_relabeled")[0]
    if len(data) == 0:
        continue
    data = data.t().reshape(-1).cpu().numpy()[1:]
    if len(data) <= 1000:
        continue
    ii_train = int(len(data)//(10/9))
    data_train = data[:ii_train]
    data_test = data[ii_train:]
    count = Counter(data_train)
    
    count_dict = {}
    accu = 0 
    for i, key in enumerate(data_train[:-1]):
        if key in count_dict:
            if data_train[i+1] in count_dict[key]:
                count_dict[key][data_train[i+1]]+=1
            else:
                count_dict[key][data_train[i+1]]=1
        else:
            count_dict[key] = {data_train[i+1]:1}
    for key, value in count_dict.items():
        if count[key]<=n:
            count_dict[key] = count.most_common(1)[0][0]
        else:
            count_dict[key] = max(count_dict[key].items(), key=operator.itemgetter(1))[0]

    for i, key in enumerate(data_test[:-1]):
        if key in count_dict:
            if count_dict[key] == data_test[i+1]:
                accu+=1
        else:
            if count.most_common(1)[0][0]== data_test[i+1]:
                accu+=1
    data_hist_5.append(accu/len(data_test))
```

```python
bins = plt.hist(data_hist_5,bins = 50)
print(f"Mean accuracy when guessing on most common given previous location: {np.mean(data_hist_5)}")
```

### Most common given depature hour

```python
# Only looking at departures
dict_tmp = {}
peps = good_peps_train
data_hist_4 = []
for pep in peps:
    try:
        with open(f"Data/{pep}/timestamps.json") as json_file:
            data_times = json.load(json_file)
        with open(f"Data/{pep}/final_series.json") as json_file:
            data_path = json.load(json_file)
    except FileNotFoundError:
        print(f"File {pep} not found")
        continue
    if len(data_path) <= 1000:
        continue
    
    data_path = [path[-1] for path in data_path]
    max_data_path = max(data_path)
    
    
    ii_train = int(len(data_path)//(10/9))
    data_path_train = data_path[:ii_train]
    data_path_test = data_path[ii_train:]
    data_times_train = data_times[:ii_train]
    data_times_test = data_times[ii_train:]
    
    
    dict_tmp[pep] = {time:{place:0 for place in range(max_data_path+1)} for time in range(24)}
    datetimes_2_train = np.array([datetime.fromtimestamp(time[1]) for time in data_times_train])
    for i,event in enumerate(datetimes_2_train):
        hour = event.hour
        dict_tmp[pep][hour][data_path[i+1]]+=1    
    for hour in dict_tmp[pep].keys():
        dict_tmp[pep][hour] = Counter(dict_tmp[pep][hour]).most_common(1)[0][0]
        
    
    datetimes_2_test = np.array([datetime.fromtimestamp(time[1]) for time in data_times_test])
    acc = 0
    count = 0
    for i,event in enumerate(datetimes_2_test):
        if i == len(datetimes_2_test)-1:
            continue
        acc += int(dict_tmp[pep][event.hour]==data_path_test[i+1])
        count += 1
    data_hist_4.append(acc/count)
        
```

```python
#Location not taken into account. So if travelling we're not looking at time differences.
bins = plt.hist(data_hist_4,bins = 50)
print(f"Mean accuracy when guessing on most common given time slot on hour basis: {np.mean(data_hist_4)}")
```

```python
# Only looking at departures
for n in range(100,200,20):
    peps = good_peps_train
    data_hist_6 = []
    for pep in peps:
        try:
            with open(f"Data/{pep}/timestamps.json") as json_file:
                data_times = json.load(json_file)
            with open(f"Data/{pep}/final_series.json") as json_file:
                data_path = json.load(json_file)
        except FileNotFoundError:
            print(f"File {pep} not found")
            continue
        if len(data_path) <= 1000:
            continue
        data_path = [path[-1] for path in data_path]
        max_data_path = max(data_path)


        ii_train = int(len(data_path)//(10/9))
        data_path_train = data_path[:ii_train]
        data_path_test = data_path[ii_train:]
        data_times_train = data_times[:ii_train]
        data_times_test = data_times[ii_train:]

        dict_tmp_hour = {time:{place:0 for place in range(max_data_path+1)} for time in range(24)}
        dict_tmp = {time:{} for time in range(24)}
        datetimes_2_train = np.array([datetime.fromtimestamp(time[1]) for time in data_times_train])
        for i,event in enumerate(datetimes_2_train):
            hour = event.hour
            dict_tmp_hour[hour][data_path[i+1]]+=1

            key = data_path[i]
            if key in dict_tmp[hour]:
                if data_path[i+1] in dict_tmp[hour][key]:
                    dict_tmp[hour][key][data_path[i+1]]+=1
                else:
                    dict_tmp[hour][key][data_path[i+1]]=1
            else:
                dict_tmp[hour][key] = {data_path[i+1]:1}  

        for hour in dict_tmp.keys():
            #dict_tmp[hour] = Counter(dict_tmp[hour]).most_common(1)[0][0]
            for loc in dict_tmp[hour].keys():
                count = Counter(dict_tmp[hour][loc])
                #for key, value in dict_tmp[hour][loc].items():
                #import pdb; pdb.set_trace()
                if count.most_common()[0][0]>=n:
                    dict_tmp[hour][loc] = Counter(dict_tmp_hour[hour]).most_common(1)[0][0]
                else:
                    dict_tmp[hour][loc] = count.most_common()[0][0]


        datetimes_2_test = np.array([datetime.fromtimestamp(time[1]) for time in data_times_test])
        acc = 0
        count = 0
        for i,event in enumerate(datetimes_2_test):
            if i == len(datetimes_2_test)-1:
                continue
            if data_path_test[i] in dict_tmp[event.hour]:
                acc += int(dict_tmp[event.hour][data_path_test[i]]==data_path_test[i+1])
            else:
                acc += int(Counter(dict_tmp_hour[event.hour]).most_common(1)[0][0]==data_path_test[i+1])
            count += 1
        data_hist_6.append(acc/count)
    print(np.mean(data_hist_6))
```

# Distribution of visits per location

```python
dist = []
one_visit_frac = []
exploration_frac = []
for pep in good_peps:
    data = load_data([pep],bs=1,dat_type="all_relabeled",stop_t=False)[0]
    if len(data)<100:
        continue
    data = data.t().reshape(-1).cpu().numpy()[1:]
    count = list(Counter(data).values())
    dist.extend(count)
    one_visit_frac.append(Counter(count)[1]/len(count))
    exploration_frac.append(len(count)/sum(count))
    
```

```python
if not entropy_dict:
    entropy_dict = {}
step = 10
for pep in tqdm(range(vali_peps)):
    if str(pep) not in entropy_dict:
        data = load_data([pep],bs=1,dat_type="all_relabeled",stop_t=False)[0]
        data = data.t().reshape(-1).cpu().numpy()[1:]
        entropy_dict[pep] = []
        for ix in range(step,len(data),step):
            entropy_dict[pep].append(utils.entropy(list(data[:ix]),tqdm_boo=True)[0])
```

```python
good_peps = []
for i,(pep,ent) in enumerate(entropy_dict.items()):
    y_slope = np.array([abs(ent[j]-ent[j+5]) for j in range(len(ent)-5)])
    for k in range(len(y_slope)-15):
        if sum(abs(y_slope[k:k+10]-y_slope[k+5:k+15])<0.05)>=8:
            good_peps.append(pep)
            break
```
